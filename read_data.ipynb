```python
# load data
import torch
train_data = torch.load('data/thws-mai-idl-ss-25-sign-language/SignLanguage_kaggle/old_annotated.pth')
print(train_data)
```

```python
# Plot a few labeled images from dataset
import matplotlib.pyplot as plt
def plot_dataset(dataset):
 print(f"Total data:{len(dataset)}")
 _, labels = dataset[:]
 print(f"label num:{len(set(labels.tolist()))}")
 fig = plt.figure(figsize=(13, 8))
 columns = 6
 rows = 3
 # ax enables access to manipulate each of subplots
 ax = []
 for i in range(columns * rows):
   img, label = dataset[i]
   # create subplot and append to ax
   ax.append(fig.add_subplot(rows, columns, i + 1))
   ax[-1].set_title("Label: " + str(label.item()))
   im = img.permute(1, 2, 0).numpy()
   plt.imshow(im)  # show image
 plt.show()  # finally, render the plot

plot_dataset(train_data)
```

```python

test_data = torch.load('data/thws-mai-idl-ss-25-sign-language/SignLanguage_kaggle/todo_example.pth')
plot_dataset(test_data)
```
